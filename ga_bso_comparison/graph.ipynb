```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
values_df = pd.read_csv('average_total_value.csv')
times_df = pd.read_csv('average_time_s.csv')

# Plotting Total Values
plt.figure(figsize=(10, 5))
values_df.plot(kind='bar', x='Filename', y=['BSO', 'GA'])
plt.title('Comparaison des valeurs totales de BSO et GA')
plt.ylabel('Valeur totale moyenne')
plt.xlabel('Instances (Benchmarks)')
plt.xticks(rotation=45)
plt.legend(title='Algorithmes')

# Plotting Execution Times
plt.figure(figsize=(10, 5))
times_df.plot(kind='line', x='Filename', y=['BSO', 'GA'], marker='o')
plt.title('Comparaison des temps d\'excution de BSO et GA')
plt.ylabel('Temps d\'excution moyen (s)')
plt.xlabel('Instances (Benchmarks)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Algorithmes')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
values_df = pd.read_csv('average_total_value.csv')
times_df = pd.read_csv('average_time_s.csv')

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Plotting Total Values
values_df.plot(kind='bar', x='Filename', y=['BSO', 'GA'], ax=ax1)
ax1.set_title('Comparaison des valeurs totales de BSO et GA')
ax1.set_ylabel('Valeur totale moyenne')
ax1.set_xlabel('Instances (Benchmarks)')
ax1.set_xticklabels(values_df['Filename'], rotation=45)
ax1.legend(title='Algorithmes')

# Plotting Execution Times
times_df.plot(kind='line', x='Filename', y=['BSO', 'GA'], marker='o', ax=ax2)
ax2.set_title('Comparaison des temps d\'excution de BSO et GA')
ax2.set_ylabel('Temps d\'excution moyen (s)')
ax2.set_xlabel('Instances (Benchmarks)')
ax2.set_xticks(range(len(times_df['Filename'])))
ax2.set_xticklabels(times_df['Filename'], rotation=45)
ax2.grid(True)
ax2.legend(title='Algorithmes')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
